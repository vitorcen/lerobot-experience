## 7. 常见问题 (FAQ)

### Q: 显存不足 (OOM) 怎么办？

1. 减小 `eval.batch_size`（例如设为 1）
2. 对于 Pi0 等大模型，确保使用高端 GPU (A100/H100) 或量化版本
3. 使用 `--policy.device=cpu`（仅作为调试，速度极慢）

### Q: EGL 渲染错误（EGL_NOT_INITIALIZED）？

这是在无头服务器上运行 LIBERO/robosuite 时的常见问题。解决方案：


````python
# 无头服务器运行示例（环境变量与命令在同一行）：
!MUJOCO_GL=egl PYOPENGL_PLATFORM=egl EGL_DEVICE_ID=0 conda run -n lerobot python lerobot/src/lerobot/scripts/lerobot_eval.py \
    --policy.path=lerobot/pi0_libero_finetuned \
    --env.type=libero \
    --env.task=libero_spatial \
    --eval.batch_size=1 \
    --eval.n_episodes=5 \
    --policy.device=cuda
````


### Q: 数据集太大，能否只下载少量 episodes 先看看？

**A**: 可以。LeRobot 支持多种方式避免一次性下载整个数据集：

**1. 流式加载 (Streaming)** — 无需下载，直接通过网络流式读取：


````python
# 数据集 >1GB，使用 chunk 精确下载 + 预览（只下该 episode 所在 chunk）
!conda run -n lerobot python scripts/download_one_episode_chunk.py lerobot/robotwin_unified --episode 0 --root ./peek_cache
````


**2. `lerobot_dataset_viz.py` 按需下载** — 可视化脚本只会下载你指定的那个 episode 的数据和视频，不会拉取整个数据集。第一次看某个 episode 时会有短暂下载，后续从缓存读取。

**3. 使用 Hugging Face `datasets` 流式 API** — 如果你只需要原始数据而不需要 LeRobot 格式：


````python
# 数据集 >1GB，使用 chunk 精确下载 + 预览（只下该 episode 所在 chunk）
!conda run -n lerobot python scripts/download_one_episode_chunk.py lerobot/robotwin_unified --episode 0 --root ./peek_cache
````


**4. 只下载 meta 信息预览** — 如果你只想快速了解数据集结构而不下载任何视频/数据：

```bash
# 克隆时跳过所有 LFS 文件（只下载元数据，通常 < 1MB）
GIT_LFS_SKIP_SMUDGE=1 git clone --depth 1 https://huggingface.co/datasets/lerobot/robotwin_unified
cat robotwin_unified/meta/info.json  # 查看 episodes、frames、fps 等统计
```

> **建议**：对于几十 GB 以上的大数据集（如 RoboCasa 184GB、DROID 380GB、RoboTwin 77GB），强烈推荐使用 **episodes 参数**或**流式加载**，避免磁盘和带宽浪费。



如果 EGL 仍然出错，尝试使用 OSMesa（CPU 软件渲染）：
